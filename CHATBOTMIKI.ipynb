```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "google/flan-t5-base"

# Load model and tokenizer with lower precision to improve performance
# Use float16 for faster inference
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name, torch_dtype=torch.float16)  
# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

print("MIKI AI Chatbot is ready! Type 'exit' to stop.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break

    prompt = f"""
You are a polite, friendly, and helpful AI assistant.
Explain concepts clearly and respectfully.

Question: {user_input}
Answer:
"""

    # Add timeout handling
    try:
        inputs = tokenizer(prompt, return_tensors="pt", truncation=True).to(device) 
        # Reduce generation parameters for faster response
        outputs = model.generate(
            **inputs,
            max_length=128,  
            temperature=0.3,
            top_p=0.9,
            repetition_penalty=1.1,
            num_beams=2,  # Simpler beam search
            early_stopping=True  # Stop generation when possible
        )

        response = tokenizer.decode(outputs[0], skip_special_tokens=True)
        print("MIKI:", response)
    except Exception as e:
        print(f"Error during generation: {e}")
        print("Try with a shorter prompt or simpler question.")
```
